```python
# ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varにランダムな整数0～100を出力
var = random.randint(0, 100)

# 変数varを出力
print(var)

# 変数varが3の倍数の場合は「Fizz」、5の倍数の場合は「Buzz」、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」、どの場合にも該当しない場合は変数の値を出力
if var%3 == 0 and var%5 == 0:
    print("FizzBuzz")
elif var%3 == 0:
    print("Fizz")
elif var%5 == 0:
    print("Buzz")
else:
     print(var)
```
